# Building the baseline classifier

We'll now do a basic round of supervised classification using scikit-learn. We start by loading the data. We actually have the final classifications in this dataset, so that we can figure out what our accuracy rate was, but we'll ignore it initially and pretend we're starting from scratch.

### Loading the data

```python
from __future__ import division, print_function

import geopandas as gpd
```

```python
roads = gpd.GeoDataFrame.from_file('input/classified_roads.geojson')
```

```python
roads
```

```python
roads["classification"][roads["road_name"]=="Kismis"] = "Malay"
roads["classification"][roads["road_name"]=="Merbok"] = "Malay"
roads["classification"][roads["road_name"]=="Onan"] = "Malay"
roads["classification"][roads["road_name"]=="Mariam"] = "Malay"
```

## Step 1: extract the raw data

We won't be actually classifying every road, but every road name - note that there are repeats of several kinds, e.g. row 34 + 35 are different sections of "Devonshire Road", there's several "Pasir Ris Drive"s with different numbers. We want to just classify each name once.

So we'll extract out the unique road names from this list.

```python
unique_roadnames = roads.groupby(["road_name", "classification"]).sum().reset_index()[["road_name", "classification"]]
unique_roadnames
```

```python
# sanity check that this is the same as the number of unique road_names: yes.

len(gpd.pd.unique(roads["road_name"]))
```

```python
# let's pick a random 10% to train with

from random import sample
train_test_set = unique_roadnames.loc[sample(unique_roadnames.index, int(len(unique_roadnames) / 10))]

X = train_test_set['road_name']
y = train_test_set['classification']
```

```python
X
```

```python
y
```

## Step 2: decide on classification schema

We want to classify roads according to linguistic origin. The most common ones are Malay, Chinese, British, Indian, Other (Ethnic Origins), and Generic. Generic includes names like Market Street and Orchard Road. While these are English words, they're not really named after British people but are descriptors of the road environs. Also, I extend names like "Orchard Road" to include other nature terms like "Chestnut Street". Although it's unlikely that chestnut trees grow there, it's still a generic nature term rather than necessarily of British origin.

Our final classification schema is:

* Malay
* Chinese
* British
* Indian
* Other
* Generic

## Step 3: decide on what features to use

What we're doing is basically language classification. Often, people use n-grams as features for this. scikit-learn conveniently provides a function that does this.

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(1,4), analyzer='char')
```

## Step 4: decide on which classifier to use

<img src="http://scikit-learn.org/stable/_static/ml_map.png">

According to this, we should be starting out with Linear SVC.

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
```

```python
from sklearn.pipeline import Pipeline

# we'll talk about what pipelines are later
pipeline = Pipeline([
    ('vect', vect),  # combine ngrams
    ('clf' , clf),   # with a classifier
])
```

## Step 5: split into train and test, classify, and evaluate!

scikit-learn provides some nice functions for splitting a dataset into training and testing. Because a single split may accidentally give you a good or bad score, we'll average over several splits of the train/test data (NB: you can also use KFolds for this).

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score

for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

So we're averaging about 50-55% accuracy on this. It's not totally terrible, because with these six categories we would be expecting a completely random classifier to have the following accuracy:

```python
import numpy as np

# accuracy if just randomly assigning labels
accuracy_if_random = 100 / 6
print(accuracy_if_random)

# accuracy while taking into account the proportions of counts
accuracy_if_guessing_proportionally = sum(np.power((y.value_counts() / len(y)), 2))
print(accuracy_if_guessing * 100)

# accuracy if just guessing the most frequent label always
accuracy_if_guessing_malay = max(y.value_counts()) / len(y)
print(accuracy_if_guessing_malay * 100)
```

But it's not terrific either. It would mean having to reclassify 50% of all the things manually. So, next step: figure out how to improve this!
