```python
import geopandas

# get a data frame of roads
roads = geopandas.GeoDataFrame.from_file('../resources/singapore-roads.geojson')

# roads.shape
# (59218, 13)
# That's a lot of roads for a small country
```

```python
roads.geometry.plot()
```

```python
# Roads from Malaysia and Indonesia are being included - even some ferry lines.

# we'll filter to just the roads within Singapore boundaries

# first let's get an outline of Singapore's administrative boundaries
admin = geopandas.GeoDataFrame.from_file('../resources/singapore-admin.geojson')

#singapore = admin.ix[0].geometry
singapore = admin.ix[0]
```

```python
geopandas.GeoSeries(singapore.geometry).plot()
```

```python
roads = roads[roads['geometry'].within(singapore.geometry)]
```

```python
roads.plot()
```

```python
roads.shape
```

```python
set(roads['class'].values)
```

```python
# store current value of roads
singapore_streets = roads
```

```python
# filter out roads with no name
roads = roads[roads['name'].notnull()]
roads.plot()
```

```python
# filter the right kinds of roads - exclude footpaths, etc

roads = roads[roads['class'] == 'highway']

# there are still things like footpaths in there.
# these are the actual road types we want

accepted_road_types = [u'motorway', u'motorway_link', u'primary', 
                       u'primary_link', u'residential', u'road',
                       u'secondary', u'secondary_link', u'tertiary', 
                       u'tertiary_link', u'trunk', u'trunk_link']

roads = roads[roads['type'].isin(accepted_road_types)]

roads.shape
```

```python
# unique road names

roadnames = set(roads['name'].values)
len(roadnames)
```

```python
roadnames
```

```python
roads.geometry.plot()
```

```python
# export both the list of roads and the filtered GeoJSON

roads.to_file('input/filtered_roads.geojson', driver='GeoJSON')
f = open('input/sg_roadnames.txt','w')
for road in sorted(roadnames):
    f.write(road)
    f.write('\n')
```
