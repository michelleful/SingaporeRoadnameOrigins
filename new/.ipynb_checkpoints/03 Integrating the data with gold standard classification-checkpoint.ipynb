## Integrate the data we got from OSM with the "gold standard" classification data

In reality I did the classification in a cyclic fashion, first hand-coding 10% of the data,
using that to train a classifier, running that on the next 10%, correcting the incorrect outputs,
using the cumulative 20% to train a classifier, and so on and so forth until 100% was hand-checked.

In this series of notebooks I'm just going to integrate the end result classification once rather
than artificially reintroducing the 10% stages.

This IPython notebook features the use of Pandas' `merge` (SQL join) command.

```python
%matplotlib inline
```

```python
import geopandas as gpd
```

```python
roads = gpd.GeoDataFrame.from_file('input/roads_for_classification.geojson')
```

```python
roads
```

```python
classification = gpd.pd.read_csv('../data/data.road_and_class.gold.csv',sep=',', header=None)
```

```python
classification.columns = ["road_name", "classification"]
classification
```

```python
classification['classification'][classification['classification']=='Nature'] = 'Generic'
```

```python
classification["road_name"] = classification["road_name"].map(str.strip)
```

```python
classification[classification['road_name']=='Ah Hood']
```

```python
# add in roads from a secondary data source for roads unclassified in the gold standard file above
# these are roads that were in OpenStreetMap but not in other map sources
# some of them are just misspellings that should be corrected in the data cleaning stage in future

other_classification = gpd.pd.read_csv('../data/data.road_and_class.extra.gold.csv',sep=',', header=None)
other_classification.columns = ["road_name", "classification"]
```

```python
final_classification = classification.append(other_classification)
```

```python
final_classification
```

```python
classified_roads = gpd.pd.merge(roads, final_classification, on='road_name', how='left')
```

```python
set(classified_roads.classification.values)
```

```python
classified_roads
```

```python
len(classified_roads)
```

```python
gpd.GeoDataFrame(classified_roads).to_file('input/classified_roads.geojson', driver='GeoJSON')
```
