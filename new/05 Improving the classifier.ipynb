# Improving the classifier

Last time, we used a very basic set of features and 10% of our dataset to train a Linear Support Vector Machine classifier, and wound up getting ~50% accuracy (accuracy of a random classifier would have been ~22%). Some ways to improve this are to:

* Increase the amount of training data
* Increase the number of features
* Decrease the number of features by doing feature selection
* Changing the classifier model, either by:
  * Using a different base classifier (e.g. RandomForests instead of LinearSVC)
  * Tuning the hyperparameters of the classifier
  
It's pretty evident how to add training data, and switch base classifiers, so we'll look at the other three in turn.

## Adding features

First we'll look at how to add your own features.

```python
# Load up all the data like we did last time

from __future__ import division, print_function

# open data in GeoPandas
import geopandas as gpd
roads = gpd.GeoDataFrame.from_file('input/classified_roads.geojson')

# manually correct some incorrect ones - should probably correct upstream, really
roads["classification"][roads["road_name"]=="Kismis"] = "Malay"
roads["classification"][roads["road_name"]=="Merbok"] = "Malay"
roads["classification"][roads["road_name"]=="Onan"] = "Malay"
roads["classification"][roads["road_name"]=="Mariam"] = "Malay"

# extract out just the road names, whether the road "tag" (i.e. Street, Jalan, etc) is Malay or not, and the classification
unique_roadnames = roads.groupby(["road_name", "classification"]).sum().reset_index()
unique_roadnames = unique_roadnames[["road_name", "has_malay_road_tag", "classification"]]

# still using 10% of the data
from random import sample
train_test_set = unique_roadnames.loc[sample(unique_roadnames.index, int(len(unique_roadnames) / 10))]
```

Before we get into adding features, let's discuss Pipelines. Pipelines are used to transform data from raw input, through data transformers like CountVectorizer, which gave us a numpy array of n-gram counts, through classifiers like LinearSVC. Why use Pipelines instead of writing functions? Basically, because we can use combine transformers and classifiers by chaining and unioning them into a single reusable (and often fairly readable) object.

Last time, we had only a single column of text that we fed through the CountVectorizer. What happens if we have multiple columns in X, only ONE of which we want to feed through CountVectorizer, as we will do in this case?

```python
X = train_test_set[["road_name", "has_malay_road_tag"]]
y = train_test_set.classification
X[:3]
```

We need to extract the column and then feed it through CountVectorizer. The following transformer code will let us select columns when passed in a numpy array

```python
class ColumnSelector(object):
    """Selects column(s) from a numpy array
    Adapted from mlxtend Machine Learning Library Extensions 
    https://github.com/rasbt/mlxtend/blob/master/mlxtend/sklearn/feature_sel.py
    """
    
    def __init__(self, cols):
        self.cols = cols
        
    def transform(self, X, y=None):
        return X[:, self.cols]

    def fit(self, X, y=None):
        return self
    
class SingleColumn(object):
    """Reshapes 1-d array (which is interpreted by sklearn as a single sample)
    to (n_samples, 1).
    """
    
    def transform(self, X, y=None):
        return X.reshape(X.size, 1)

    def fit(self, X, y=None):
        return self

# I only really need this because of the way I extracted the roads from Pandas, using sum
import numpy as np
class BooleanBinarizer(object):
    """Given array X, returns array X' where value of an elt is 1 if its underlying evaluates to True, 0 otherwise
    """
    
    def transform(self, X, y=None):
        return X.astype(np.bool).astype(np.int64)

    def fit(self, X, y=None):
        return self
    
class MakeDtype(object):
    def transform(self, X, y=None):
        return X.astype(np.float64)

    def fit(self, X, y=None):
        return self
    
```

```python
from sklearn.pipeline import Pipeline, FeatureUnion

from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(1,4), analyzer='char')

from sklearn.svm import LinearSVC
clf = LinearSVC()

pipeline = Pipeline([
    ('extract_roadname', ColumnSelector(cols=(0))),  # extracts roadname column
    ('vect', vect),  # extract ngrams from roadnames
    ('clf' , clf),   # feed the output through a classifier
])
```

```python
# This should be functionally equivalent to our previous model - and so it is

from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!

print(classification_report(y_test, y_true))

print(confusion_matrix(y_test, y_true))
```

```python
# now we also want to use the feature of whether the road had a Malay road "tag" 
# (Jalan, Lorong, etc) rather than an English one (Road, Street, etc)

# but these are sort of parallel feature extraction processes
# Pipeline combines feature extractions in serial mode 
# in order to do the parallel combination, we have to do a FeatureUnion

ngram_pipeline = pipeline = Pipeline([
    ('extract_roadname', ColumnSelector(cols=(0))),  # extracts roadname column
    ('vect', vect),  # extract ngrams from roadnames
])

malay_roadtag_pipeline = Pipeline([
    ('extract_malay_roadtag', ColumnSelector(cols=(1))),  # extracts Malay road tag column
    ('single_column', SingleColumn()),
    ('binarizer', BooleanBinarizer())
])

pipeline = Pipeline([
    ('features', FeatureUnion([
        ('ngram_pipeline', ngram_pipeline),
        ('malay_roadtag_pipeline', malay_roadtag_pipeline),
    ])),
    ('clf' , clf),   # feed the output through a classifier
])
```

```python
for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

```python
# more features!

# (3) number of words. More words usually correlates well with Chinese names.

# instead of inheriting from object, can also inherit from TransformerMixin

from sklearn.base import TransformerMixin
class Apply(TransformerMixin):
    """Applies a function f element-wise to the numpy array
    """
    
    def __init__(self, fn):
        self.fn = np.vectorize(fn)
        
    def transform(self, X, y=None):
        return self.fn(X)

    def fit(self, X, y=None):
        return self
```

```python
# starting here, I'll introduce a less verbose way of putting together pipelines and feature unions

from sklearn.pipeline import make_pipeline, make_union

# with the functions make_pipeline and make_union, you don't need to (and can't) name the individual steps
num_words_pipeline = make_pipeline(
    ColumnSelector(cols=(0)),
    SingleColumn(),
    Apply(lambda roadname: len(roadname.split())),
)

pipeline = make_pipeline(
    make_union(
        ngram_pipeline,
        malay_roadtag_pipeline,
        num_words_pipeline,
    ),
    clf,
)

for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

```python
# more features!

# (4) average length of word. Short words correlates with Chinese names, long ones with Indian names.

ave_word_length_pipeline = make_pipeline(
    ColumnSelector(cols=(0)),
    SingleColumn(),
    Apply(lambda roadname: np.mean([len(w) for w in roadname.split()]))
)

pipeline = make_pipeline(
    make_union(
        ngram_pipeline,
        malay_roadtag_pipeline,
        num_words_pipeline,
        ave_word_length_pipeline,
    ),
    clf,
)

for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

```python
# more features!

# (5) are all the words in the roadname in the dictionary?
# if yes, it's likely to be Generic
# of course, there are outliers like Chin Chew Road

# first we need a dictionary

from glob import glob
dictionary = dict()
for dictfile in glob('../resources/scowl-7.1/final/english-words*'):
    if dictfile.endswith('95'):
        continue
    with open(dictfile, 'r') as g:
        for line in g.readlines():
            dictionary[line.strip()] = 1
```

```python
from operator import and_

all_words_in_dict_pipeline = make_pipeline(
    ColumnSelector(cols=(0)),
    SingleColumn(),
    Apply(lambda roadname: 1 if reduce(and_, [word.lower() in dictionary
                                              for word in roadname.split()]) else 0)
)

feature_pipeline = make_union(
        ngram_pipeline,
        malay_roadtag_pipeline,
        num_words_pipeline,
        ave_word_length_pipeline,
        all_words_in_dict_pipeline
)


pipeline = make_pipeline(
    feature_pipeline,
    clf,
)


for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

Another way we could have done this was to calculate these features in our Pandas dataframe, and then just used ColumnSelector extract out the relevant columns and done only a single FeatureUnion on the ngrams and the rest. ## Feature selection

Although we've been looking at n-grams kind of as a single feature, it's really thousands of features. Having so many features introduces the risk of over-fitting the model. We can introduce a feature selector into the pipeline, before the classifier, to select only the most important features. I only experimented with two: SelectPercentile and running Linear SVC in L1 penalty mode.

```python
from sklearn.feature_selection import SelectPercentile, chi2

pipeline = make_pipeline(
    feature_pipeline,
    SelectPercentile(chi2, percentile=50),  # select the 50% highest scoring features
    clf,
)


for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

```python
pipeline = make_pipeline(
    feature_pipeline,
    LinearSVC(C=1, penalty="l1", dual=False),
    clf,
)


for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```

```python
from sklearn.decomposition import PCA

class MakeDenseArray(TransformerMixin):
    """Converts sparse array to dense array"""
    
    def transform(self, X, y=None):
        return X.toarray()

    def fit(self, X, y=None):
        return self

pipeline = make_pipeline(
    feature_pipeline,
    MakeDenseArray(),
    PCA(),
    clf,
)

for i in range(5):
    X_train, X_test, y_train, y_true = train_test_split(X, y)
    classifier = pipeline.fit(X_train, y_train)  # train the classifier!
    y_test = classifier.predict(X_test)          # apply it to the test data!
    print(accuracy_score(y_test, y_true))        # compare the results to the "ground truth" / "gold standard" data!
```
