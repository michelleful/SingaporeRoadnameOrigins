```python
import geopandas as gpd

# get a data frame of roads
roads = gpd.GeoDataFrame.from_file('../resources/singapore-roads.geojson')
```

```python
# See how many features (roads) there are and how many parameters each has
roads.shape
```

```python
# 59218 is a lot of roads for a small country...let's plot them to see what's going on
```

```python
roads.geometry.plot() # NB time-consuming to plot - avoid rerunning
```

```python
# Roads from Malaysia and Indonesia are being included - even some ferry lines.

# we'll filter to just the roads within Singapore boundaries

# first let's get an outline of Singapore's administrative boundaries
admin = gpd.GeoDataFrame.from_file('../resources/singapore-admin.geojson')
```

```python
singapore = admin.ix[0]
```

```python
gpd.GeoSeries(singapore.geometry).plot()
```

```python
sg_roads = roads[roads.geometry.within(singapore.geometry)]
```

```python
sg_roads.shape
```

```python
set(sg_roads["class"].values)
```

```python
# filter out roads with no name
sg_roads = sg_roads[sg_roads.name.notnull()]
```

```python
sg_roads.plot()
```

```python
# filter the right kinds of roads - exclude footpaths, etc

sg_roads = sg_roads[sg_roads["class"] == 'highway']

# there are still things like footpaths in there.
# these are the actual road types we want

accepted_road_types = [u'motorway', u'motorway_link', u'primary', 
                       u'primary_link', u'residential', u'road',
                       u'secondary', u'secondary_link', u'tertiary', 
                       u'tertiary_link', u'trunk', u'trunk_link']

sg_roads = sg_roads[sg_roads["type"].isin(accepted_road_types)]
```

```python
# remove exit ramps
sg_roads = sg_roads[~sg_roads.name.str.contains('Exit')]

sg_roads.shape
```

```python
# unique road names

len(sg_roads['name'].unique())
```

```python
sg_roads.geometry.plot()
```
