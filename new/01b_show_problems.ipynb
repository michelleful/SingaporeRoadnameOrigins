```python
import geopandas as gpd

# get a data frame of roads
roads = gpd.GeoDataFrame.from_file('../resources/singapore-roads.geojson')
```

```python
# See how many features (roads) there are and how many parameters each has
roads.shape
```

```python
# 59218 is a lot of roads for a small country...let's plot them to see what's going on
```

```python
roads.geometry.plot() # NB time-consuming to plot - avoid rerunning
```

```python
# Roads from Malaysia and Indonesia are being included - even some ferry lines.

# we'll filter to just the roads within Singapore boundaries

# first let's get an outline of Singapore's administrative boundaries
admin = gpd.GeoDataFrame.from_file('../resources/singapore-admin.geojson')
```

```python
# METHOD 1 for obtaining the Singapore geometry

singapore = admin.ix[0]
```

```python
gpd.GeoSeries(singapore.geometry).plot()
```

```python
sg_roads = roads[roads.geometry.within(singapore.geometry)]
```

```python
sg_roads.shape
```
